```python
lista_de_graus = []

for i in range(1, 20+1):

    graus = {}

    with open(f'lista_de_pares_amostra_{i}.dat', 'r') as f:
        for linha in f:
            vertice1, vertice2 = linha.strip().split()
            graus[vertice1] = graus.get(vertice1, 0) + 1
            graus[vertice2] = graus.get(vertice2, 0) + 1

    lista_de_graus.append(graus.values())

lista_de_todos_os_graus = [grau for sublist in lista_de_graus for grau in sublist]

import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(figsize=(10, 6), ncols=5, nrows=4)
axs = axs.flatten()
for amostra, lista in enumerate(lista_de_graus):
    sns.histplot(lista, bins=30, kde=False, ax=axs[amostra])
    axs[amostra].set_title(f'Amostra {amostra+1}')

#plt.tight_layout()
#plt.show()

sns.histplot(lista_de_todos_os_graus, bins=100, kde=False)
```
